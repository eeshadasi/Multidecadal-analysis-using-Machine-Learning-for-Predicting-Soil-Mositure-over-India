```python
import os
import numpy as np
import pandas as pd
import scipy.io as sio

data_folder = 'Data'
output_folder = 'output6'
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

yearly_array_data = []
year_to_store = 2020
years_latitude_data = {}
years_longitude_data = {}
nan_indices_starting_year = {}

days_in_months = {
    'Jan': 31, 'Feb': 29, 'Mar': 31, 'Apr': 30, 'May': 31, 'Jun': 30,
    'Jul': 31, 'Aug': 31, 'Sep': 30, 'Oct': 31, 'Nov': 30, 'Dec': 31
}

month_in_days = {
    'Jan': 1, 'Feb': 2, 'Mar': 3, 'Apr': 4, 'May': 5, 'Jun': 6,
    'Jul': 7, 'Aug': 8, 'Sep': 9, 'Oct': 10, 'Nov': 11, 'Dec': 12
}

def days_in_month(year, month):
    if month == 'Feb':
        if year % 4 == 0 and (year % 100 != 0 or year % 400 == 0):
            return 29
        else:
            return 28
    else:
        return days_in_months[month]

for filename in os.listdir(data_folder):
    if filename.startswith('sm') and filename.endswith('.mat'):
        year_str = filename[2:6]
        if not year_str.isdigit():
            continue
        year = int(year_str)
        filepath = os.path.join(data_folder, filename)
        mat_data = sio.loadmat(filepath)
        sm_data = mat_data['sm']
        yearly_array_data.append({'Year': year, 'Data': sm_data})

df_list = []
nan_indices = []

for year_data in yearly_array_data:
    year = year_data['Year']
    data = year_data['Data']
    nan_indices_starting_year[year] = []  # Initialize the list for storing NaN indices

    latitude = np.linspace(0, data.shape[0] - 1, data.shape[0])
    longitude = np.linspace(0, data.shape[1] - 1, data.shape[1])
    latitude_grid, longitude_grid = np.meshgrid(latitude, longitude)
    years_latitude_data[year] = latitude_grid
    years_longitude_data[year] = longitude_grid
    if year == year_to_store:
        year_latitude_data = latitude_grid
        year_longitude_data = longitude_grid

    for month_idx, month in enumerate(days_in_months.keys()):
        days = days_in_month(year, month)
        month_data = np.nanmean(data[:, :, month_idx * days:(month_idx + 1) * days], axis=2)

        for i in range(len(latitude_grid)):
            for j in range(len(longitude_grid)):
                value = month_data[i][j]
                latitude_val = latitude_grid[i][j]
                longitude_val = longitude_grid[i][j]
                if np.isnan(value):
                    nan_indices_starting_year[year].append((len(df_list),'Value'))
                    nan_indices.append((len(df_list), 'Value'))
                    value = np.nanmean(month_data)
                df_list.append({
                    'Year': year,
                    'Month': month_in_days[month],
                    'Value': value,
                    'Latitude': longitude_val,
                    'Longitude': latitude_val
                })
df = pd.DataFrame(df_list)
df2=df.copy()

for idx, column in nan_indices:
    df2.at[idx, 'Value'] = np.nan
start_longitude = 68.125
start_latitude = 5.375
longitude_step = 0.25
latitude_step = 0.25
df['Longitude'] = start_longitude + df['Longitude'] * longitude_step
df['Latitude'] = start_latitude + df['Latitude'] * latitude_step
```
